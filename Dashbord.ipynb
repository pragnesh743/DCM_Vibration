```python
def plot_fig(df,x,y,value_):
    
    fig = go.Figure()
#     df=df[:value_]
    # Add traces
    fig.add_trace(go.Scatter(
        x=df[x],
        y=df[y],
        name=y,
        yaxis="y",
    ))

    fig.update_layout(hovermode='x unified')
    fig.update_layout(legend=dict(
            yanchor="top",
            y=0.99,
            xanchor="left",
            x=0.01
        ),
        margin=dict(l=20, r=20, t=20, b=20),
        paper_bgcolor="#E8EEF2",
    )
    return fig
```

```python
def plotAnomaly(th, th_type, data,col_pos):
    actual = th.iloc[:,col_pos]
    max_th = actual.max()
    min_th = actual.min()
    if th_type == 'high':
        percent = (max_th*20)/100
        marker = max_th + percent
    elif th_type == 'low':
        percent = (max_th*20)/100
        marker = min_th - percent
    elif th_type == 'posTh':
        percent = (max_th*10)/100
        marker = max_th + percent
    elif th_type == 'negTh':
        percent = (max_th*10)/100
        marker = min_th - percent
    # print(marker,th_type,maxTh,minTh)
    columnname = list(data.columns.values.tolist()) 
    cond = {'true': True, 'false': False}
    anomaly = data.loc[(data.iloc[:,col_pos] == cond['true']), [columnname[col_pos]]]
    anomaly[columnname[col_pos]] = marker
    # print(anomaly)
    return anomaly
```

```python

def Plot_prediction_anomaly(actual,prediction,high,low,posTh,negTh,prm):
# Plot actual and predicted value
    i=i=list(actual.columns).index(prm)
    print(i)
    colnames = actual.columns
    fig1 = go.Figure()
    
    pred = prediction.copy()

    colname = colnames[i]
    # columnname = colname.split('_')
    fig1.add_trace(
        go.Scattergl(name=actual.columns[i],x=actual.iloc[:,i].index,y=actual.iloc[:,i],mode="lines"))
    fig1.add_trace(
        go.Scattergl(name=pred.columns[i],x=pred.iloc[:,i].index,y=pred.iloc[:,i],mode="lines"))
    
    
#     axe[i].plot(actual.iloc[:,i].index,actual.iloc[:,i], color='blue',label="Actual")
#     axe[i].plot(pred.iloc[:,i].index,pred.iloc[:,i], color='green',label="Predicted")

    # Get anomaly index values
    anomaly_high = plotAnomaly(th = actual , th_type = 'high', data = high , col_pos = i)
    anomaly_low = plotAnomaly(th = actual , th_type = 'low', data = low , col_pos = i)
    anomaly_posR = plotAnomaly(th = actual , th_type = 'posTh', data = posTh , col_pos = i)
    anomaly_negR = plotAnomaly(th = actual , th_type = 'negTh', data = negTh , col_pos = i)

    # Plot the anomalies
    fig1.add_trace(
        go.Scattergl(x=anomaly_high[anomaly_high.columns[0]].index,y=anomaly_high[anomaly_high.columns[0]],mode="markers",
                     marker=dict(color='red'),
                     showlegend=False
                    ))
    fig1.add_trace(
        go.Scattergl(x=anomaly_low[anomaly_low.columns[0]].index,y=anomaly_low[anomaly_low.columns[0]],mode="markers",
                     marker=dict(color='red'),
                     showlegend=False
                    ))
    fig1.add_trace(
        go.Scattergl(x=anomaly_posR[anomaly_posR.columns[0]].index,y=anomaly_posR[anomaly_posR.columns[0]],mode="markers",
                     marker=dict(color='yellow'),
                     showlegend=False
                    ))
    fig1.add_trace(
        go.Scattergl(x=anomaly_negR[anomaly_negR.columns[0]].index,y=anomaly_negR[anomaly_negR.columns[0]],mode="markers",
                     marker=dict(color='yellow'),
                     showlegend=False
                 ))
    # align title
    fig1.update_layout(title_text=actual.columns[i].split('_actual')[0], title_x=0.5)
    return fig1
```
